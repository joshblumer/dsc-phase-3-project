```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from scipy import stats
```

```python
df_train = pd.read_csv('experimental2_training_set')
df_train.head()
```

```python
df_train['lga'].value_counts()
```

```python
def top_lgas(var):
    if var in df_train['lga'].value_counts(normalize = True).head(25).index:
        return var
    else:
        return 'other'
```

```python
df_train['top_lga'] = df_train['lga'].map(top_lgas)
```

```python
df_train.drop(['lga'], axis = 1, inplace = True)
```

```python
df_train['top_lga'].value_counts()
```

```python
df_train['top_lga'].replace('Lushoto', 'other', inplace = True)
```

```python
df_train['top_funded'].replace('Amref', 'other', inplace = True)
```

```python
df_train['scheme_management'].replace('None', 'other', inplace = True)
```

```python
df_train['top_installers'].replace('CES', 'other', inplace = True)
```

```python
df_train.to_csv('experimental2_training_set', index = False)
```
